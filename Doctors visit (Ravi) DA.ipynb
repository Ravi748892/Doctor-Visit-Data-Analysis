```python
# % of the people getting govt Insurance due to low income
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting govt health Insurance due to low income ")
plt.show()

# % of the people having private Insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of the people having private Insurance ")
plt.show()

# % of the people getting govt Insurance due to  old age, disability or veteran status
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("%  of the people getting govt Insurance due to  old age, disability or veteran status ")
plt.show()
```

```python
db= df.groupby('gender')['reduced'].sum().to_frame().reset_index()
# creating a bar chart
plt.barh(db['gender'],db['reduced'],color = ['cornflowerblue','lightseagreen'])
# adding the aesthetics
plt.title('Bar Chat')
plt.xlabel('gender')
plt.ylabel('reduced activity')
# show the plot
plt.show()
```
